```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/Dominion/Documents/england_premier_league_players_data.csv")
data.head()
```

```python
c1 = data[(data["club"]=="Arsenal")]
c2 = data[(data["club"]=="Chelsea")]
c3 = data[(data["club"]=="Liverpool")]
x = c1.fpl_points.sum()
y = c2.fpl_points.sum()
z = c3.fpl_points.sum()
```

```python
amount = [x,y,z]
label = ["Total FPL Points by Arsenal Players", "Total FPL Points by Chelsea Players", "Total FPL Points by Liverpool players"]
```

```python
plt.pie(amount, labels = label, colors =("red", "blue", "yellow"),autopct = "%1.1f%%",shadow = True, startangle = 90)
plt.title("Premier League Clubs Total FPL points")
plt.show()
```
